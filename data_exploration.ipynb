# Data Preprocessing & Cleaning 🧹

This notebook demonstrates the data processing pipeline for the **Lead Scoring System**. 

Following the requirements in the `README.md`, we handle:
1. **Missing Values**: Imputing numerical data with the median and categorical data with mode/defaults.
2. **Text Normalization**: Cleaning strings to ensure consistency across the system.

```python
import pandas as pd
import numpy as np
from data_processing import handle_missing_values, clean_text_data, preprocess_data

# Create messy sample data for demonstration
data = {
    'Lead_ID': ['L001', 'L002', 'L003', 'L004'],
    'Industry': ['Technology ', ' Healthcare', np.nan, 'Finance'],
    'Company Size': ['11-50', '51-200', '11-50', np.nan],
    'Website Visits': [15, np.nan, 5, 20],
    'Pricing Page Time': [10, 2, np.nan, 8]
}

df_messy = pd.DataFrame(data)
print("--- Messy Data (Before) ---")
display(df_messy)
```

## 1. Handling Missing Values

As specified in the project requirements:
- **Numerical Columns**: Filled with the **median** value of the column.
- **Industry**: Filled with **'Unknown'**.
- **Company Size**: Filled with the **mode** (most frequent value).

```python
df_imputed = handle_missing_values(df_messy.copy())
print("--- Data After Missing Value Imputation ---")
display(df_imputed)
```

## 2. Text Normalization

Consistency is key for scoring. We ensure all text is:
- **Lowercase**: To avoid duplicates like 'Tech' vs 'tech'.
- **Stripped**: Removing leading/trailing whitespace.

```python
df_cleaned = clean_text_data(df_imputed)
print("--- Data After Text Cleaning ---")
display(df_cleaned)
```

## Summary

By using the `preprocess_data` function, we can transform any raw lead data into a structured format ready for our lead scoring model.

```python
# Final check with the main preprocessing function
final_df = preprocess_data(pd.DataFrame(data))
print("--- Final Preprocessed Data ---")
display(final_df)
```
